```python
import os
import numpy as np
from collections import OrderedDict

dataset_root         = r"C:\Users\21654\Downloads\dataset_again"
train_static_npz     = os.path.join(dataset_root, "test_static.npz")
train_sequences_npz  = os.path.join(dataset_root, "test_sequences.npz")
static_dir    = os.path.join(dataset_root, "test", "static")
sequences_dir = os.path.join(dataset_root, "test", "sequences")


def extract_ordered_label_indices(static_npz, seq_npz):
    data_s = np.load(static_npz, mmap_mode="r")
    data_q = np.load(seq_npz,   mmap_mode="r")
    y_static = data_s["y"]
    y_seq    = data_q["y"]
    y_all    = np.concatenate([y_static, y_seq], axis=0)
    # ordered unique
    return list(OrderedDict.fromkeys(y_all.tolist()))

label_indices = extract_ordered_label_indices(train_static_npz,
                                              train_sequences_npz)
print("Indices de classes, dans l'ordre :", label_indices)


classes_static = sorted(
    d for d in os.listdir(static_dir)
    if os.path.isdir(os.path.join(static_dir, d))
)
classes_seq = sorted(
    d for d in os.listdir(sequences_dir)
    if os.path.isdir(os.path.join(sequences_dir, d))
)
all_classes = sorted(set(classes_static) | set(classes_seq))
print("Toutes les classes (triées)   :", all_classes)


class_names = [all_classes[i] for i in label_indices]
NUM_CLASSES = len(class_names)

print("class_names ordonné        :", class_names)
print("Nombre de classes (NUM)    :", NUM_CLASSES)
```
